## 2. Montagem do manifesto local

Como a pasta atual contem apenas imagens, vamos montar um manifesto simples com `image_path` e `image_id`.

```python
def build_align_manifest(dataset_root, max_samples=None):
    dataset_root = Path(dataset_root)
    images = sorted(dataset_root.glob('*.jpg'))
    if not images:
        raise FileNotFoundError('Nenhuma imagem .jpg foi encontrada em img_align_celeba.')

    if max_samples:
        images = images[:max_samples]

    rows = []
    for index, image_path in enumerate(images):
        rows.append({
            'image_id': index,
            'image_name': image_path.name,
            'image_path': str(image_path),
        })
    return pd.DataFrame(rows)
```

```python
manifest_df = build_align_manifest(DATASET_ROOT, max_samples=10)
manifest_df.head()
```

```python
manifest_output = ARTIFACTS_DIR / 'celeba_align_manifest.parquet'
full_manifest_df = build_align_manifest(DATASET_ROOT)
full_manifest_df.to_parquet(manifest_output, index=False)
manifest_output
```
