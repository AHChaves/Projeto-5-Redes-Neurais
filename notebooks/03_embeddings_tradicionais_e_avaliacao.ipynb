# 03 - Embeddings Tradicionais e Avaliacao

Neste notebook geramos o baseline tradicional com `HOG` e comparamos os resultados.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve().parent
SRC_DIR = PROJECT_ROOT / 'src'

if str(SRC_DIR) not in sys.path:
    sys.path.append(str(SRC_DIR))
```

```python
import pandas as pd

from projeto_redes_neurais.embeddings.traditional import extract_hog_embeddings
from projeto_redes_neurais.evaluation.visualization import reduce_embeddings, save_scatter_plot
from projeto_redes_neurais.models.classical import train_classical_model
```

```python
DATASET_ROOT = Path('/caminho/para/CelebA-Spoof')
OUTPUT_PATH = PROJECT_ROOT / 'artifacts' / 'embeddings' / 'train' / 'traditional_hog.parquet'
MAX_SAMPLES = 200
```

## Extracao dos embeddings HOG

```python
hog_df = extract_hog_embeddings(
    dataset_root=DATASET_ROOT,
    split='train',
    output_path=OUTPUT_PATH,
    max_samples=MAX_SAMPLES,
)
hog_df.head()
```

## Projecao para visualizacao

```python
hog_2d = reduce_embeddings(hog_df, method='pca')
hog_2d.head()
```

```python
figure_path = PROJECT_ROOT / 'reports' / 'figures' / 'traditional_hog_pca.png'
save_scatter_plot(hog_2d, figure_path, 'HOG Embeddings - PCA')
figure_path
```

## Modelo classico sobre embeddings HOG

```python
result = train_classical_model(hog_df, model_name='svm')
print('Acuracia:', round(result.accuracy, 4))
print('Macro F1:', round(result.macro_f1, 4))
print(result.report)
```
